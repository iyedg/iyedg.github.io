```python
def age_groups_data(start_date: str = None, end_date: str = None) -> pd.DataFrame:

    date_format = "%Y-%m-%dT%H:%M:%S.%fZ"

    start_date = start_date.strftime(date_format)
    o_end_date = end_date
    end_date = end_date.strftime(date_format)
    timestamp_date = datetime.today().strftime(date_format)

    payload = {
        "batch": [
            {
                "request": {
                    "params": {
                        "index": "evax_inscription_open_data",
                        "body": {
                            "aggs": {
                                "3": {
                                    "range": {
                                        "field": "age_inscrit",
                                        "ranges": [
                                            {"to": 18, "from": 0},
                                            {"to": 30, "from": 18},
                                            {"to": 40, "from": 30},
                                            {"to": 50, "from": 40},
                                            {"to": 60, "from": 50},
                                            {"to": 75, "from": 60},
                                            {"from": 75},
                                        ],
                                        "keyed": True,
                                    },
                                    "aggs": {"2": {"sum": {"field": "convoque"}}},
                                }
                            },
                            "size": 0,
                            "fields": [
                                {"field": "@timestamp", "format": "date_time"},
                                {"field": "date_inscrit", "format": "date_time"},
                                {"field": "last_update_time", "format": "date_time"},
                            ],
                            "script_fields": {},
                            "stored_fields": ["*"],
                            "runtime_mappings": {},
                            "_source": {"excludes": []},
                            "query": {
                                "bool": {
                                    "must": [],
                                    "filter": [
                                        {"match_all": {}},
                                        {"match_phrase": {"id_etat_inscrit": 1}},
                                        {
                                            "range": {
                                                "date_inscrit": {
                                                    "gte": start_date,
                                                    "lt": end_date,
                                                    "format": "strict_date_optional_time",
                                                }
                                            }
                                        },
                                        {
                                            "range": {
                                                "@timestamp": {
                                                    "gte": "2021-06-20T17:12:54.491Z",
                                                    "lte": timestamp_date,
                                                    "format": "strict_date_optional_time",
                                                }
                                            }
                                        },
                                    ],
                                    "should": [],
                                    "must_not": [],
                                }
                            },
                        },
                        "preference": 1626800226037,
                    }
                },
                "options": {"strategy": "ese"},
            }
        ]
    }
    response = s.post(
        "https://opendata.evax.tn/s/evax/internal/bsearch",
        headers=headers,
        json=payload,
        verify=False,
    )
    raw_data = response.json()["result"]["rawResponse"]["aggregations"]["3"]["buckets"]
    formatted_data = []
    for value in raw_data.values():
        invited = value["2"]["value"]
        registered = value["doc_count"]
        age_from = value.get("from") or ""
        age_to = value.get("to") or ""
        formatted_data.append(
            {
                "age_group": f"{age_from}-{age_to}",
                "invited": invited,
                "registered": registered,
                "date": o_end_date.date(),
            }
        )
    return pd.DataFrame(formatted_data)
```

```python
def time_series():
    date_format = "%Y-%m-%dT%H:%M:%S.%fZ"
    timestamp_date = datetime.today().strftime(date_format)
    payload = {
        "batch": [
            {
                "request": {
                    "params": {
                        "index": "evax_inscription_open_data",
                        "body": {
                            "aggs": {
                                "2": {
                                    "date_histogram": {
                                        "field": "date_inscrit",
                                        "fixed_interval": "24h",
                                        "time_zone": "Africa/Tunis",
                                    },
                                    "aggs": {
                                        "1": {
                                            "cumulative_sum": {"buckets_path": "_count"}
                                        }
                                    },
                                }
                            },
                            "size": 0,
                            "fields": [
                                {"field": "@timestamp", "format": "date_time"},
                                {"field": "date_inscrit", "format": "date_time"},
                                {"field": "last_update_time", "format": "date_time"},
                            ],
                            "script_fields": {},
                            "stored_fields": ["*"],
                            "runtime_mappings": {},
                            "_source": {"excludes": []},
                            "query": {
                                "bool": {
                                    "must": [],
                                    "filter": [
                                        {"match_all": {}},
                                        {
                                            "range": {
                                                "@timestamp": {
                                                    "gte": "2021-01-01T14:30:20.952Z",
                                                    "lte": timestamp_date,
                                                    "format": "strict_date_optional_time",
                                                }
                                            }
                                        },
                                    ],
                                    "should": [],
                                    "must_not": [],
                                }
                            },
                        },
                        "preference": 1626792302617,
                    }
                },
                "options": {"strategy": "ese"},
            }
        ]
    }

    response = s.post(
        "https://opendata.evax.tn/s/evax/internal/bsearch",
        headers=headers,
        json=payload,
        verify=False,
    )
    try:
        return pd.json_normalize(
            response.json()["result"]["rawResponse"]["aggregations"]["2"]["buckets"]
        ).drop(columns=["1.value", "key"])
    except KeyError:
        return response.json()
```

```python
time_series()
```

```python
ts_df = time_series().pipe(
    lambda df: df.assign(
        key_as_string=pd.to_datetime(df["key_as_string"]),
        rolling=df["doc_count"].rolling(5).mean(),
        cumsum=df["doc_count"].cumsum()
    )
)
```

```python
by_age_dfs = []
for d1, d2 in zip(ts_df["key_as_string"][:-1], ts_df["key_as_string"][1:]):
    try:
        by_age_dfs.append(age_groups_data(start_date=d1, end_date=d2))
    except KeyError:
        print(d1, d2)
```

```python
agg_df = pd.concat(by_age_dfs)
```
